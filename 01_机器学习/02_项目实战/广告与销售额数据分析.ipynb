```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1 加载数据
ad_spend = np.array([5,10,15,20,25,30,35,40,45,50,55,60,65,70,75])
sales= np.array([22,40,58,75,92,110,128,145,162,180,198,215,233,250,268])
```

```python
#  2.数据集拆分（占比：训练集 70% 测试集30%）
# 训练集  目前设定为前十组，模型在训练的时候看不见
x_train = ad_spend[:10]
y_train = sales[:10]
# 训练集  目前设定为后5组
x_test = ad_spend[10:]
y_test = sales[10:]
plt.scatter(x_train,y_train,label='train',color='royalblue',s=60)
plt.scatter(x_test,y_test,label='test',color='crimson',s=60)
plt.xlabel('guanggao')
plt.ylabel('xiaoshoue')
plt.legend()
plt.title('guanggaoe&xiaoshoue of anlasy')
plt.show()
```
